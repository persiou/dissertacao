```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rankdata

ACCENT = "#117738"      # cor de destaque
N_PTS = 100              # numero de pares observados
SEED = 2
rng = np.random.default_rng(SEED)

# Relacao monotonica nao-linear em forma de sigmoide
x = rng.uniform(-6, 6, N_PTS)
y = 1 / (1 + np.exp(-x)) + rng.normal(0, 0.03, N_PTS)

rx = rankdata(x)        # postos de x
ry = rankdata(y)        # postos de y

plt.rcParams.update({"font.family": "Arial", "mathtext.fontset": "stixsans", "axes.titlesize": 10, "axes.labelsize": 9, "xtick.labelsize": 9, "ytick.labelsize": 9})

cm = 1/2.54
fig, axes = plt.subplots(1, 2, figsize=(15*cm, 5.5*cm), layout="constrained")

# ----------------------------------------------------------------------
# Painel (a): valores originais
# ----------------------------------------------------------------------
ax = axes[0]
ax.scatter(x, y, c='gray', s=12, alpha=0.8, edgecolors='black', linewidth=0.2)
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_title("(a)")

# ----------------------------------------------------------------------
# Painel (b): postos das variaveis
# ----------------------------------------------------------------------
ax = axes[1]
ax.scatter(rx, ry, c=ACCENT, s=12, alpha=0.8, edgecolors='black', linewidth=0.2)
ax.set_xlabel("ranks de $x$")
ax.set_ylabel("ranks de $y$")
ax.set_title("(b)")

fig.savefig("figuras/exemplo_spearman.png", dpi=300)
```
