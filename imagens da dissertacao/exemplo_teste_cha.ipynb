```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from math import comb

mpl.rcParams.update({"font.family": "Arial", "mathtext.fontset": "stixsans", "axes.labelsize": 9, "xtick.labelsize": 9, "ytick.labelsize": 9})
cm = 1/2.54

# Distribuição hipergeométrica do número de acertos sob H0
k = np.arange(0, 5)                      # acertos entre as 4 xícaras de um tipo
acertos = 2 * k                          # total de acertos entre as 8 xícaras
prob = np.array([comb(4, ki)**2 for ki in k]) / comb(8, 4)

fig, ax = plt.subplots(figsize=(12*cm, 6*cm), layout="constrained")

cores = ['gray'] * len(acertos)
cores[-1] = "#099E42"
ax.bar(acertos, prob, width=1.0, color=cores, edgecolor='black', linewidth=0.3)

# Anotação do valor-p sobre a barra de acerto total
ax.annotate(f'$p \\approx {prob[-1]:.4f}$',
            xy=(8, prob[-1]), xytext=(8, prob[-1] + 0.08),
            ha='center', fontsize=9,
            arrowprops=dict(arrowstyle='->', linewidth=0.7))

ax.set_xlabel('Xícaras corretamente classificadas')
ax.set_ylabel('Probabilidade')
ax.set_xticks(acertos)
ax.set_ylim(0, prob.max() + 0.15)


plt.savefig('figuras/exemplo_teste_cha.png', dpi=300)
```
