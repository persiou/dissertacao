```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

centers = [(-1.5, 1.5), (1.5, 1.5), (-1.5, -1.5), (1.5, -1.5)]
X, _ = make_blobs(n_samples=300, centers=centers, cluster_std=0.7, random_state=42)

# Partição pelo k-means
labels = KMeans(n_clusters=4, n_init=10, random_state=2).fit_predict(X)


mpl.rcParams.update({"font.family": "Arial", "mathtext.fontset": "stixsans", "axes.titlesize": 10, "axes.labelsize": 10, "xtick.labelsize": 9, "ytick.labelsize": 9})

cm = 1/2.54
fig, axes = plt.subplots(1, 2, figsize=(15*cm, 5.5*cm), layout="constrained")

axes[0].scatter(X[:, 0], X[:, 1], c='gray', s=8, alpha=0.7, edgecolors='black', linewidth=0.2)
axes[0].set_xlabel('$X_1$')
axes[0].set_ylabel('$X_2$')
axes[0].set_title('(a)')

axes[1].scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=8, alpha=0.7, edgecolors='black', linewidth=0.2)
axes[1].set_xlabel('$X_1$')
axes[1].set_ylabel('$X_2$')
axes[1].set_title('(b)')

plt.savefig('figuras/exemplo_agrupamento.png', dpi=300)
```
