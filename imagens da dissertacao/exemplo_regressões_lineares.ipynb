```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt


rng = np.random.default_rng(2)
n = 40
x = np.linspace(0, 10, n)

beta0, beta1 = -1.0, 0.5
y = beta0 + beta1 * x + rng.normal(0, 0.5, n)


# Ajuste aos dados criados
b1, b0 = np.polyfit(x, y, 1)
y_hat = b0 + b1 * x

mpl.rcParams.update({"font.family": "Arial", "mathtext.fontset": "stixsans", "axes.labelsize": 10, "xtick.labelsize": 9, "ytick.labelsize": 9})

cm= 1/2.54
fig, ax = plt.subplots(figsize=(10*cm, 5*cm), layout="constrained")

# Indicação dos resíduos com os traços verticais
for xi, yi, yhi in zip(x, y, y_hat):
    ax.plot([xi, xi], [yi, yhi], color='gray', linewidth=0.8, alpha=0.7, zorder=1)


ax.scatter(x, y, c='gray', s=8, alpha=0.8, edgecolors='black', linewidth=0.2, zorder=2)
ax.plot(x, y_hat, color="#099E42", linewidth=2, zorder=3)

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

plt.savefig('figuras/exemplo_regressao_linear.png', dpi=300)
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt


rng = np.random.default_rng(2)
n = 40
x = np.linspace(0.5, 10, n)

beta0, beta1 = -1.0, 4.5
# Relação logarítmica com erro aditivo bem comportado
y = beta0 + beta1 * np.log(x) + rng.normal(0, 0.5, n)


mpl.rcParams.update({"font.family": "Arial", "mathtext.fontset": "stixsans", "axes.titlesize": 10, "axes.labelsize": 10, "xtick.labelsize": 9, "ytick.labelsize": 9})

cm= 1/2.54
fig, axes = plt.subplots(1, 2, figsize=(15*cm, 6*cm), layout="constrained")

# Painel (a): escala original, relação curva
b1, b0 = np.polyfit(x, y, 1)
axes[0].scatter(x, y, c='gray', s=8, alpha=0.8, edgecolors='black', linewidth=0.2)
axes[0].plot(x, b0 + b1 * x, color="#099E42", linewidth=1.5)
axes[0].set_xlabel('$x$')
axes[0].set_ylabel('$y$')
axes[0].set_title('(a)')

# Painel (b): x transformado, relação linear
xt = np.log(x)
b1, b0 = np.polyfit(xt, y, 1)
axes[1].scatter(xt, y, c='gray', s=8, alpha=0.8, edgecolors='black', linewidth=0.2)
axes[1].plot(xt, b0 + b1 * xt, color="#099E42", linewidth=1.5)
axes[1].set_xlabel('$\\log x$')
axes[1].set_ylabel('$y$')
axes[1].set_title('(b)')

plt.savefig('figuras/exemplo_regressao_transformacao.png', dpi=300)
```
